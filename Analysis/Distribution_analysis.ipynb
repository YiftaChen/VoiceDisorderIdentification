```python
import matplotlib.pyplot as plt
import seaborn as sns
import json
import pandas as pd


# %matplotlib qt
```

## Male Analysis

```python
def sum_pathologies(path_tree):
    return {pathology:sum(sub_pathologies.values()) for (pathology,sub_pathologies) in path_tree.items()}


def create_pie_chart(kv_dict,title=None, size="big"):
    # #define data
    data = kv_dict.values()
    labels = kv_dict.keys()
    # labels = ['label1', 'label2', 'label3']

    # #define Seaborn color palette to use
    # custom_palette = sns.color_palette("Paired", 9)

    colors = sns.color_palette('Paired',13)
    colors.pop(5)
    # #create pie chart
    if size == "big":
        plt.figure(figsize=(8, 6), dpi=160)
    else:
        plt.figure(figsize=(4, 3), dpi=80)
    if title != None:
        plt.title(title)
    def my_autopct(pct):
        return ('%.0f%%' % pct) if pct >= 1 else ''
    plt.pie(data, labels = labels, colors = colors, autopct=my_autopct)
    plt.show()

```

```python
male = {}
with open("male_pathologies.json",'r') as f:
    male = json.load(f)

female = {}
with open("female_pathologies.json",'r') as f:
    female = json.load(f)

both = {}
with open("both_genders_pathologies.json",'r') as f:
    both = json.load(f)


male_summed = sum_pathologies(male)
female_summed = sum_pathologies(female)
both_summed = sum_pathologies(both)

```

```python
male = {}
with open("male_pathologies.json",'r') as f:
    male = json.load(f)
male_summed = sum_pathologies(male)
```

```python
create_pie_chart(male_summed)
```

```python
for pathology,sub_pathologies in male.items():
#     print(sub_pathologies)
    create_pie_chart(sub_pathologies, title=f"{pathology} male",size="small")
```
